```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from matplotlib import pyplot as plt

df = pd.read_csv("exam_dataset.csv")

print("\nSummary statistics:")
display(df.describe())

if df["Gender"].dtype == object:
    df["Gender"] = df["Gender"].map({"Female": 0, "Male": 1})

print("\nUpdated 'Gender' column values:")
print(df["Gender"].value_counts())

X = df[["HoursStudied", "PreviousScore", "ParentalInvolvement", "Gender"]]
y = df["ExamScore"]

# Add constant for OLS regression
X = sm.add_constant(X)


#RUN MULTIPLE LINEAR REGRESSION
model = sm.OLS(y, X).fit()
print("\nRegression Summary:")
print(model.summary())

# VISUALIZATION 1: Previous Score vs Exam Score
plt.scatter(df["PreviousScore"], df["ExamScore"])
plt.xlabel("Previous Score")
plt.ylabel("Exam Score")
plt.title("Previous Score vs Exam Score")
plt.grid(True)
plt.show()

#VISUALIZATION 2: Actual vs Predicted Exam Scores
pred = model.predict(X)

plt.scatter(df["ExamScore"], pred)
plt.xlabel("Actual Exam Score")
plt.ylabel("Predicted Exam Score")
plt.title("Actual vs Predicted Exam Scores")
plt.plot(
    [df["ExamScore"].min(), df["ExamScore"].max()],
    [df["ExamScore"].min(), df["ExamScore"].max()]
)
plt.grid(True)
plt.show()

```
