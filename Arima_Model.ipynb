```r

# Load libraries
library(forecast) #Runs ARIMA models, forecasting, ACF/PACF, model selection
library(tseries) #Provides ADF test to check stationarity
library(ggplot2) #Used for cleaner visualizations

# Step 2: Defining data set
data <- read.csv("/content/full_cleaned_dataset.csv")

# Step 3: Model selection (ARIMA model)
arima_model <- auto.arima(gdp_ts)

# Step 4: View model result
# Display the summary of the selected ARIMA model (shows orders, coefficients, etc.)
summary(arima_model)

# Step 5: Check model diagnostic
# Check residuals for white noise (plots ACF, histogram, and Q-Q plot)
checkresiduals(arima_model)

# Step 6: Generate forecast
# Generate forecasts for the next 8 quarters (adjust h as needed)
forecasts <- forecast(arima_model, h = 8)

# Print the forecast results (point forecasts and confidence intervals)
print(forecasts)

# Step 7: Plot the forecast
# Visualize the historical data, fitted model, and forecasts
autoplot(forecasts) +
  ggtitle("Philippine Quarterly GDP: Historical and ARIMA Forecast") +
  xlab("Year") +
  ylab("GDP (Billions PHP)") +
  theme_minimal()
```
